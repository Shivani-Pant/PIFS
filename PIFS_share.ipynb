Creation of PIFS and fuzzification based on the clusters obtained from ARC in above step:

```python
a=13000
b=20000
print(a,b)
n=len(center)
h=(b-a)/n                              # Step length of interval where n represents number of clusters obtained
l=[0.0]*n                              # defining lower bound of fuzzy sets
mi=[0.0]*n                             # defining middle of fuzzy sets
u=[0.0]*n                              # defining upper bound of fuzzy sets
for i in range(n):
  l[i]=a+i*h                           # UOD=[a,b]
  mi[i]=a+(i+1)*h
  u[i]=a+(i+2)*h
  if (u[i]>b):
    u[i]=b
for i in range(n):
  print('Fuzzy set %d is:'%(i+1),'[%s,%s,%s]'%(l[i],mi[i],u[i]))
arr=[[0.0]*3]*n
for i in range(n):
  arr[i]=[l[i],mi[i],u[i]]
std1=[0.0]*n
mean1=[0.0]*n
for i in range(n):
  mean1[i]=statistics.mean(arr[i])
  std1[i]=statistics.pstdev(arr[i])
memb=[ [0 for i in range(n)] for j in range(len(ar)) ]  # defining membership matrix corresponding to fuzzy sets
for j in range(n):
  for i in range(len(ar)):
    if (l[j]<=ar[i]<=mi[j]):                 # ar[i] is the ith time series observation
      memb[i][j]=(ar[i]-l[j])/(mi[j]-l[j])
    if (mi[j]<ar[i]<=u[j]):
      memb[i][j]=(u[j]-ar[i])/(u[j]-mi[j])
    if ar[i]<l[j] or ar[i]>u[j]:
      memb[i][j]=0
for i in range(len(ar)):
  print(memb[i])
print('-------------------------------------------transpose of triangular membership-------------------------------------------------------------------')
newmatrix=[ [memb[j][i] for j in range(len(ar))] for i in range(n)]
for j in range(n):
  print(newmatrix[j])
print('----------------------------------------------IFS membership----------------------------------------------------------------')
e=[0.0]*n
f=[0.0]*n
mu=[0.0]*n
nmu=[0.0]*n
for i in range(n):
  mat1=[]
  for j in range(len(ar)):
    if newmatrix[i][j]!=0:
      mat1.append(newmatrix[i][j])
  if len(mat1)!=0:
    e[i]=max(x for x in newmatrix[i] if x!=0)
    f[i]=min(x for x in newmatrix[i] if x!=0)
    mu[i]=e[i]*f[i]
    nmu[i]=1-mu[i]
  else:
    e[i]=0
    f[i]=0
    mu[i]=0
    nmu[i]=0
mem=[ [0 for j in range(len(ar))] for i in range(n)]
nonmem=[ [0 for j in range(len(ar))] for i in range(n)]
pi=[ [0 for j in range(len(ar))] for i in range(n)]
for i in range(n):
  for j in range(len(ar)):
    if newmatrix[i][j]!=0:
      mem[i][j]=round(newmatrix[i][j]*nmu[i],6)
      nonmem[i][j]=round(1-mu[i]-mem[i][j],6)
      pi[i][j]=round(1-mem[i][j]-nonmem[i][j],6)
    else:
      mem[i][j]=0
      nonmem[i][j]=0
      pi[i][j]=0
for i in range(n):
  print(mem[i])
print('--------------------------------------------IFS non-membership------------------------------------------------------------------')
for i in range(n):
  print(nonmem[i])
print('---------------------------------------------IFS degree of hesitancy-----------------------------------------------------------------')
for i in range(n):
  print(pi[i])
print('----------------------------------------------Probability of IFS----------------------------------------------------------------')
Prob=[ [0 for j in range(len(ar))] for i in range(n)]
nProb=[ [0 for j in range(len(ar))] for i in range(n)]
for i in range(n):
  for j in range(len(ar)):
    exp1=0.0
    exp2=0.0
    if mem[i][j]>0:
      exp1=math.exp(-((ar[j]-((mem[i][j]-1)*h)-mean1[i])**2)/(2*std1[i]*std1[i]))
      exp2=math.exp(-((ar[j]-((-mem[i][j]+1)*h)-mean1[i])**2)/(2*std1[i]*std1[i]))
      Prob[i][j]=(h/((math.sqrt(2*(math.pi)))*std1[i]))*(exp1+exp2)
      nProb[i][j]=1-Prob[i][j]
    else:
      Prob[i][j]=0.0
      nProb[i][j]=0.0
for i in range(n):
  print(Prob[i])
print('------------------------------------------------Non-Probability of IFS-------------------------------------------------------------------')
for i in range(n):
  print(nProb[i])
imem=[ [0 for j in range(len(ar))] for i in range(n)]
inmem=[ [0 for j in range(len(ar))] for i in range(n)]
for i in range(n):
  for j in range(len(ar)):
      if mem[i][j]>0:
        imem[i][j]=1-math.pow(1-mem[i][j],Prob[i][j])
        inmem[i][j]=1-math.pow(nonmem[i][j],nProb[i][j])
nimem=[ [imem[j][i] for j in range(n)] for i in range(len(ar))]
ninmem=[ [inmem[j][i] for j in range(n)] for i in range(len(ar))]
for i in range(len(ar)):
  print('final memb=',nimem[i])
for i in range(len(ar)):
  print('final nonmemb=',ninmem[i])
ind=[0.0]*len(ar)
for i in range(len(ar)):
    ind[i]=nimem[i].index(max(nimem[i]))
print('------------------------------------------------fuzzification--------------------------------------------------------------------')
for i in range(len(ar)):
  print(ind[i]+1)

```

```python
for i in range(n):
  print('Membership in triangular fuzzy set %d:'%(i+1),end='  ')
  for j in range(len(ar)):
    print('<',ar[j],',',newmatrix[i][j],'>',end='')
  print('\n')
print('*****************The PIFS is given by:*********************')
for i in range(n):
  print('PIFS %d:'%(i+1))
  for j in range(len(ar)):
    print('<',ar[j],',',mem[i][j],'|',Prob[i][j],',',nonmem[i][j],'|',nProb[i][j],'>',end='')
  print('\n')
print('******************Aggregated PIFS is given by:*************')
for i in range(n):
  print('Aggregated PIFS %d:'%(i+1),end='')
  for j in range(len(ar)):
    print('<',ar[j],',',imem[i][j],',',inmem[i][j],'>',end='')
  print('\n')
```

```python
tp=[ [0.0 for j in range(n)] for i in range(n)]         # intializing transition probability matrix
for i in range(len(ar)):
  if i+1<len(ar):
    k=ind[i]
    for j in range(n):
      if ind[i+1]==j:
        tp[k][j]+=1
for i in range(n):
  print(tp[i])
sum=[0.0]*n
for i in range(n):
  for j in range(n):
    sum[i]+=tp[i][j]
for i in range(n):
  for j in range(n):
    if sum[i]>0:
      tp[i][j]=tp[i][j]/sum[i]
print('--------------------------------The transition probability Matrix is:-----------------------------------------------------')
for i in range(n):
  print(tp[i])
print(tp[0][1])
```

Enrolment forecast

```python
mid=[0.0]*n
for i in range(n):
  mid[i]=(mi[i]+l[i])/2
print('The middle of intervals is:\n',mid)
pre=[0.0]*(len(ar)+1)
#for i in range(1,len(ar)+1):
for i in range(1,len(ar)+1):
  k=ind[i-1]
  for j in range(n):
    if j<k or j>k:
      pre[i]+=mid[j]*tp[k][j] #mid[j]
    if j==k:
      pre[i]+=tp[k][k]*ar[i-1]
  print('before this we get:',pre[i])
  if pre[i]==0:
    pre[i]=mid[k]
  if i==len(ar):
    pre[i]=pre[i]+(arra[i]-ar[i-1])
  if i!=len(ar):
    pre[i]=pre[i]+(ar[i]-ar[i-1])
print('The forecasted output is:')
for i in range(len(ar)+1):
  print(pre[i])
```

```python
se=0.0
mse=0.0
ts=0.0
MAD=0.0
MAPE=0.0
for i in range(1,len(ar)):
  se+=(ar[i]-pre[i])**2
  ts+=(ar[i]-pre[i])
  MAD+=abs(ar[i]-pre[i])
  MAPE+=abs(ar[i]-pre[i])/ar[i]
mse=se/(len(ar)-1)
MAD=MAD/(len(ar)-1)
ts=ts/MAD
MAPE=(MAPE/(len(ar)-1))*100
print(mse)
print('RMSE=',math.sqrt(mse))
print('Tracking Signal=',ts,'MAD=',MAD,'MAPE=',MAPE)
```
